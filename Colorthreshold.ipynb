```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread('pizza_bluescreen.jpg')
print(image.shape)
type(image)
```

```python
plt.imshow(image)
```

```python
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
lower_blue = np.array([0,0,200])

upper_blue = np.array([250,250,255])
```

Create the mask

```python
mask = cv2.inRange(image_copy, lower_blue, upper_blue)
plt.imshow(mask, cmap='gray')
```

```python
masked_image = np.copy(image_copy)

masked_image[mask!=0] = [0,0,0]

plt.imshow(masked_image)
```

```python
background_image = cv2.imread('space_background.jpg')
image = cv2.cvtColor(background_image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
background_image.shape
```

```python
crop_background = image[0:514, 0:816]
plt.imshow(crop_background)
```

```python
crop_background[mask == 0] = [0,0,0]
```

```python
plt.imshow(crop_background)
```

```python
complete_image = masked_image + crop_background
plt.imshow(complete_image)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread('rabit.jpeg')
img_copy = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_copy)
img_copy.shape
```

```python
lower_green = np.array([0,200,0])
upper_green = np.array([250,255,250])
```

```python
mask = cv2.inRange(img_copy, lower_green, upper_green)
plt.imshow(mask, cmap='gray')
```

```python
masked_image = np.copy(img_copy)

masked_image[mask!=0] = [0,0,0]

plt.imshow(masked_image)
```

```python
back_image = cv2.imread('space_background.jpg')
plt.imshow(back_image)
back_image.shape
```

```python
croped_img = back_image[0:480, 0:852]
cropped_img = cv2.cvtColor(croped_img, cv2.COLOR_BGR2RGB)
print(croped_img.shape)
plt.imshow(cropped_img)
```

```python
cropped_img[mask == 0] = [0,0,0]
plt.imshow(cropped_img)
```

```python
final_image = masked_image + cropped_img
plt.imshow(final_image)
```
