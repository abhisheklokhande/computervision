Importing useful libraries



```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import matplotlib.image as mpimg
```

Reading an image using mpimg function and converting the image to gray using cvtColor function of cv2 library

```python
image = mpimg.imread('face.jpeg')
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image, cmap='gray') 
```

Displaying Specific pixel value 

```python
x = 115
y = 400
print(gray_image[y,x])
```

Displaying maximum and minimum grayscale value

```python
np.amax(gray_image)
```

```python
np.amin(gray_image)
```

Displaying shape of image

```python
gray_image.shape
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python
ima = mpimg.imread('wa_state_highway.jpg')
plt.imshow(ima)
```

Isolating RGB channels

```python
r = ima[:,:,0]
g = ima[:,:,1]
b = ima[:,:,2]
```

```python
b
```

Visualize the individual color channel

```python
f, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,10))

ax1.set_title('R')
ax1.imshow(r, cmap='gray')

ax2.set_title('g')
ax2.imshow(g, cmap='gray')

ax3.set_title('b')
ax3.imshow(b, cmap='gray')
```
